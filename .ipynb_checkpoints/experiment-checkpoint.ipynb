```python
from utils import load_train
from utils import load_test

from model import SRCNN
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
image_size = 33
stride = 14
scale = 3
```

```python
X_train, Y_train = load_train(image_size=image_size, stride=stride, scale=scale)
```

```python
# View random input image
n = np.random.randint(100)
plt.subplot(121)
plt.title('X_train')
plt.imshow(X_train[n].reshape(image_size,image_size), cmap='gray')
plt.subplot(122)
plt.title('Y_train')
plt.imshow(Y_train[n].reshape(image_size,image_size), cmap='gray')
```

```python
X_pre_test, X_test, Y_test = load_test(scale=scale)
```

```python
n = 0 #np.random.randint(5)
plt.subplot(131)
plt.title('X_pre_test')
plt.imshow(X_pre_test[n].reshape(170,170), cmap='gray')
plt.subplot(132)
plt.title('X_test')
plt.imshow(X_test[n].reshape(510,510), cmap='gray')
plt.subplot(133)
plt.title('Y_test')
plt.imshow(Y_test[n].reshape(510,510), cmap='gray')
```

```python
c_dim=1
learning_rate=0.001
batch_size=64
epochs=100
```

```python
srcnn = SRCNN(
        image_size=image_size,
        c_dim=c_dim,
        is_training=True,
        learning_rate=learning_rate,
        batch_size=batch_size,
        epochs=epochs)
```

```python
models = srcnn.get_model_fns()
results = {x: None for x in models}
```

```python
for name,build in models.items():
    build(name)
    results[name] = srcnn.train(X_train[:1000],Y_train[:1000])
```

```python
def plot_graphs(history, string, legend=None):
    plt.plot(history.history[string])
    # plt.plot(history.history['val_'+string], '')
    plt.xlabel("Epochs")
    plt.ylabel(string)
    # if (legend):
    #     plt.legend(legend)
    # else:
    #     plt.legend([string, 'val_'+string])
    # plt.show()
```

```python
for n,h in results.items():
    plt.title('10k training imgs')
    plot_graphs(h,'psnr')
plt.legend(results.keys())
plt.show
```
